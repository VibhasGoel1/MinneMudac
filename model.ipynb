```python
# Define a function to categorize the values based on the range
def categorize(value):
    if 0 <= value < 20:
        return '0-20'
    elif 20 <= value < 40:
        return '20-40'
    elif 40 <= value < 60:
        return '40-60'
    elif 60 <= value < 80:
        return '60-80'
    else:
        return '80-100'

# Apply the function to the 'Cumm_home_game_win%' column and create the new column
df['Cumm_Home_game_Cat'] = df['Cumm_home_game_win%'].apply(categorize)
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('Attendance', axis=1) 
y = df['Attendance'].copy()    
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = .25, random_state=0)
train_X.shape, test_X.shape, train_y.shape, test_y.shape 
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn import set_config

set_config(display='diagram') # shows the pipeline structure graphically

num_pipeline = Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ])
```

```python
from sklearn.preprocessing import OneHotEncoder
cat_pipeline = Pipeline([
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("cat_encoder", OneHotEncoder(sparse=False, handle_unknown='ignore'))
    ])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_selector, make_column_transformer
import numpy as np


num_attributes = ['Visiting team game number','Home team game number','Year', 'Month','counter_visit_allstar','counter_home_allstar','Capacity', 'Total Score','Accesibility Score']

cat_attributes = ['Number of game','Day of week','Visiting team','Visiting team league','Home team', 'Home team league','Day/night','is_rivalry']


preprocess_pipeline = make_column_transformer(
    (num_pipeline, num_attributes),
    (cat_pipeline, cat_attributes),
)
preprocess_pipeline
```

```python

# List of columns you want to visualize
columns = ['Visiting team game number', 'Home team game number', 'Year', 'Month', 'counter_visit_allstar',
           'counter_home_allstar', 'Capacity', 'Total Score', 'Accesibility Score']

# Determine the layout for subplots
num_rows = 3
num_cols = 4

# Create the figure and subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 15))

# Loop through the columns and create histograms
for index, column in enumerate(columns):
    row = index // num_cols
    col = index % num_cols
    sns.histplot(data=df, x=column, ax=axes[row, col], kde=True)
    axes[row, col].set_title(f'Distribution of {column}')

# Remove unused subplots
for index in range(len(columns), num_rows * num_cols):
    row = index // num_cols
    col = index % num_cols
    fig.delaxes(axes[row][col])

# Add padding between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Preprocess the data using the defined pipelines
train_X_prepared = preprocess_pipeline.fit_transform(train_X)
test_X_prepared = preprocess_pipeline.transform(test_X)


# Train the Random Forest model
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(n_estimators=100, random_state=0)
forest_reg.fit(train_X_prepared, train_y)

# Make predictions
train_y_pred = forest_reg.predict(train_X_prepared)
test_y_pred = forest_reg.predict(test_X_prepared)

# Evaluate the model using metrics such as Mean Squared Error (MSE) and R-squared
from sklearn.metrics import mean_squared_error, r2_score

train_mse = mean_squared_error(train_y, train_y_pred)
train_rmse = np.sqrt(train_mse)
train_r2 = r2_score(train_y, train_y_pred)

test_mse = mean_squared_error(test_y, test_y_pred)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(test_y, test_y_pred)

print("Train RMSE:", train_rmse)
print("Train R-squared:", train_r2)

print("Test RMSE:", test_rmse)
print("Test R-squared:", test_r2)
```

**Assumptions and characteristics to consider when using a Random Forest Regressor**:

Non-linear relationships: Random Forests can capture complex, non-linear relationships between features and the target variable, unlike linear regression models that assume a linear relationship.

No need for feature scaling: Random Forests do not require feature scaling because they are not sensitive to the scale of input features. Decision trees split the data based on feature values, so the relative scale of the features does not impact the model's performance.

Robust to multicollinearity: Random Forests are less affected by multicollinearity among features, as the algorithm's random selection of features for each tree reduces the impact of highly correlated features on the overall model.

Model interpretability: While individual decision trees are relatively easy to interpret, Random Forest models can be harder to interpret due to the ensemble nature of the algorithm. However, feature importances can be calculated to understand the relative importance of different features in the model.

Overfitting: Random Forests can sometimes overfit the data, especially when the trees are very deep or the dataset is very small. To avoid overfitting, you can control the depth of the trees, use a higher number of estimators (trees), or perform cross-validation.

Computational complexity: Random Forests can be computationally expensive, especially when the number of trees or the depth of trees is high. You can control the computational cost by adjusting the number of trees, tree depth, and parallelizing the training process using the n_jobs parameter.

Randomness: The algorithm relies on randomness to create diverse trees. The performance of the model can vary depending on the random seed used during training. To reduce the impact of randomness, you can increase the number of trees in the ensemble or try different random seeds and average the results.

```python

feature_importances = forest_reg.feature_importances_

# Combine the column names from the preprocessing pipelines (numeric and categorical)
num_attribs = list(train_X[num_attributes])
cat_encoder = preprocess_pipeline.named_transformers_['pipeline-2']['cat_encoder']
cat_attribs = list(cat_encoder.get_feature_names_out(cat_attributes))


attributes = num_attribs + cat_attribs

# Display feature importances with corresponding column names
importances = sorted(zip(feature_importances, attributes), reverse=True)


import pandas as pd
importances_df = pd.DataFrame(importances, columns=['Importance', 'Feature'])

```

```python
importances_30 = importances_df.head(30)
importance_values = importances_30['Importance'].values
feature_names = importances_30['Feature'].values
plt.bar(feature_names,importance_values)
plt.xticks(rotation=90);
```

Feature importance interpretation :

1. Importance does not imply causality: Feature importance scores only indicate the contribution of a feature to the model's performance; they do not imply a causal relationship between the feature and the target variable. Causal inference requires more rigorous statistical analysis or experimental design.
2. Bias towards high cardinality features: MDI-based feature importance can be biased towards features with a high number of distinct values (high cardinality), such as continuous features or categorical features with many categories. These features tend to have more potential split points, and as a result, they may appear more important than they actually are.
3. Correlated features: When two or more features are highly correlated, one of them may capture most of the importance, while the others receive significantly reduced importance scores. This may not necessarily mean that the other correlated features are unimportant. In such cases, consider exploring feature selection techniques like recursive feature elimination or regularization to handle correlated features.
4. Interpretation of importance values: The absolute value of feature importance is not directly interpretable. It's the relative importance of the features that matters. A higher importance value for a feature indicates that it contributes more to the model's predictive performance compared to features with lower importance values. Comparing the importance scores can help you understand which features are more influential in the model's decision-making process.

```python

import xgboost as xgb

xgb_reg = xgb.XGBRegressor(objective ='reg:squarederror', n_estimators=100, seed=0)


xgb_reg.fit(train_X_prepared, train_y)

train_y_pred_xgb = xgb_reg.predict(train_X_prepared)
test_y_pred_xgb = xgb_reg.predict(test_X_prepared)


from sklearn.metrics import mean_squared_error, r2_score

train_mse_xgb = mean_squared_error(train_y, train_y_pred_xgb)
train_rmse_xgb = np.sqrt(train_mse_xgb)
train_r2_xgb = r2_score(train_y, train_y_pred_xgb)

test_mse_xgb = mean_squared_error(test_y, test_y_pred_xgb)
test_rmse_xgb = np.sqrt(test_mse_xgb)
test_r2_xgb = r2_score(test_y, test_y_pred_xgb)

print("Train RMSE (XGBoost):", train_rmse_xgb)
print("Train R-squared (XGBoost):", train_r2_xgb)

print("Test RMSE (XGBoost):", test_rmse_xgb)
print("Test R-squared (XGBoost):", test_r2_xgb)
```

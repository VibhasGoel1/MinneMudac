```python
import pandas as pd
import glob

# Create a list of file paths for all text files in the shared drive folder
file_paths = glob.glob('/content/drive/Shareddrives/Minnie Mudac/MinneMUDAC2023/VotingData/sportsref_download*.xls')

# Create an empty list to hold the dataframes for each file
dfs = []

# Initialize year and counter
year = 1990
counter = 0

# Loop through the list of file paths and read each file into a dataframe
for file_path in file_paths:
    tables = pd.read_html(file_path, header=1)  # Use the 2nd row (index 1) as the header
    if tables and len(tables) > 0:
        df = tables[0].iloc[:10]  # Select only rows from the 2nd row (index 1) to the 12th row (index 11)
        df['Year'] = year  # Add a new column called 'Year' and assign the same year value to all rows
        dfs.append(df)
        counter += 1  # Increment counter after processing each file
        if counter % 2 == 0:  # Increment the year after processing every two files
            year += 1
            if year == 2020:  # Skip the year 2020
                year += 1

# Concatenate the list of dataframes into a single dataframe
df_all = pd.concat(dfs, axis=0, ignore_index=True)
```
